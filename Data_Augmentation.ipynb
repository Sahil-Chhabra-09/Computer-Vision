#### Augmentation on a single image

```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import load_img
```

```python
img = load_img('/content/drive/MyDrive/My Images for CV/Sahil/WIN_20230424_21_41_49_Pro.jpg', target_size = (400,500))
```

```python
img
```

```python
type(img)
```

```python
# performing data augmentation

datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    # vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
)
```

```python
# in opencv it is directly loaded as numpy array
import numpy as np

img_np = np.array(img)
```

```python
img_np.shape
```

```python
input = img_np.reshape(1, 400, 500, 3)
```

Generating 10 randomly augmented images

```python
i = 0

# for folder, we use datagen.flow_from_directory
for output in datagen.flow(input, batch_size = 1, save_to_dir = '/content/drive/MyDrive/My Images for CV/aug'):
  i = i+1
  if i==10:
    break
```

```python
import os
```

```python
dir = "/content/drive/MyDrive/My Images for CV/aug"
aug_images = []
for files in os.listdir(dir):
  aug_img = load_img(dir + '/' + str(files), target_size = (400,500))
  aug_images.append(aug_img)
```
